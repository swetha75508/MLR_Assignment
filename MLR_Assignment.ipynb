## Data Preprocessing

## a. Handling Missing Values:
If any columns have missing data, decide whether to fill them or drop those rows/columns.

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

## b. Categorical Encoding:
Convert categorical features such as FuelType into numeric values using one-hot encoding.

```python
# One-hot encoding for 'FuelType'
df = pd.get_dummies(df, columns=['Fuel_Type'], drop_first=True)
```

```python
df
```

## c. Scaling Features:
Scale continuous variables to standardize them, especially for Lasso and Ridge.



```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['age', 'KM', 'HP', 'cylinder_volume', 'Weight']] = scaler.fit_transform(df[['age', 'KM', 'HP', 'cylinder_volume', 'Weight']])
```

```python
df
```

## Step 2: Splitting the Dataset
Now that your data is preprocessed, split it into training and testing sets.

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = df.drop(['Price'], axis=1)
y = df['Price']

# Split data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

## Step 3:  Building Multiple Linear Regression Models
Youâ€™ll now build multiple linear regression models using sklearn. We will build 3 different models by varying the input features.

 ## Model 1: Using All Features

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the model
model1 = LinearRegression()
model1.fit(X_train, y_train)

# Predict on test set
y_pred1 = model1.predict(X_test)

# Model evaluation
print("R-squared:", r2_score(y_test, y_pred1))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred1)))

```

 ## Model 2: Removing Correlated Features
From the correlation heatmap, remove one or two highly correlated features. For example, if CC and HP are highly correlated, remove one.

```python
X_train2 = X_train.drop(['cylinder_volume'], axis=1)
X_test2 = X_test.drop(['cylinder_volume'], axis=1)

# Train model
model2 = LinearRegression()
model2.fit(X_train2, y_train)

# Predict and evaluate
y_pred2 = model2.predict(X_test2)
print("R-squared:", r2_score(y_test, y_pred2))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred2)))
```

## Model 3: Using Only Important Features
From EDA, use only the features that have a strong correlation with Price (e.g., Age, KM, HP).

```python
# Select only important features
X_train3 = X_train[['age', 'KM', 'HP']]
X_test3 = X_test[['age', 'KM', 'HP']]

# Train model
model3 = LinearRegression()
model3.fit(X_train3, y_train)

# Predict and evaluate
y_pred3 = model3.predict(X_test3)
print("R-squared:", r2_score(y_test, y_pred3))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred3)))
```

## Step 4: Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Assuming you have a trained model called 'model'
y_pred = model1.predict(X_test)
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Print out the evaluation metrics
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
```

## Step 5: Lasso and Ridge Regression
Now, apply Lasso and Ridge regression, which add regularization to prevent overfitting.

## Lasso Regression
Lasso shrinks the coefficients of less important variables to zero, effectively performing feature selection.

```python
from sklearn.linear_model import Lasso

# Train Lasso regression model
lasso = Lasso(alpha=0.1)  # Alpha is the regularization strength
lasso.fit(X_train, y_train)

# Predict and evaluate
y_pred_lasso = lasso.predict(X_test)
print("Lasso R-squared:", r2_score(y_test, y_pred_lasso))
print("Lasso RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lasso)))
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge

# Train Ridge regression model
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

# Predict and evaluate
y_pred_ridge = ridge.predict(X_test)
print("Ridge R-squared:", r2_score(y_test, y_pred_ridge))
print("Ridge RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_ridge)))
```

And standardizing the data also reduces the computation time if we have a big dataset.We can use StandardScaler to standardize our data.

## 2.What techniques can be used to address multicollinearity in multiple linear regression.
To address multicollinearity in multiple linear regression, we can use the following techniques:

Remove highly correlated predictors: Identify and eliminate variables that are strongly correlated with each other.
    
Principal Component Analysis (PCA): Reduce the dimensionality of the data by transforming variables into uncorrelated components.
    
Ridge Regression (L2 regularization): Adds a penalty to the size of the coefficients, reducing multicollinearity impact.
    
Lasso Regression (L1 regularization): Shrinks some coefficients to zero, effectively performing feature selection.
    
Variance Inflation Factor (VIF): Identify and remove variables with high VIF values to reduce multicollinearity.
